```python
import pandas as pd
```

```python
def calculate_rating_classification(rating):
    return {
        rating <= 2     : 'низкий рейтинг',
        2 < rating <= 4 : 'средний рейтинг',
        rating > 4      : 'высокий рейтинг'
    } [True]

ratings = pd.read_csv('ratings.csv')
ratings['class'] = ratings['rating'].apply(calculate_rating_classification)

ratings.head()
```

```python
geo_data = {
'Центр': ['моск', 'тул', 'ярославл', 'ярославс'],
'Северо-Запад': ['петербург', 'псков', 'мурманск'],
'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']
}
```

```python
def add_geo_data(row):
    for key, value in geo_data.items():
        for x in value:
            if x in row:
                return key
    return 'undefined'
```

```python
queries = pd.read_csv('keywords.csv')
queries['region'] = queries['keyword'].apply(add_geo_data)
```
